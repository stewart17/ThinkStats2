## Exercises 1.2

Write code that reads the respondent file, 2002FemResp.dat.gz. Cross-validate the respondent and pregnancy files by comparing pregnum for each respondent with the number of records in the pregnancy file.

```python
import numpy as np

#Using packages written by the book author to read stata file and create Maps from dataframe to validate counts
import nsfg
import thinkstats2

#Using thinkstat2 module written by author read respondent stata file and convert it into data frame

respdct_file = '2002FemResp.dct'
respdat_file= '2002FemResp.dat.gz'
respdct = thinkstats2.ReadStataDct(respdct_file)
respdf = respdct.ReadFixedWidth(respdat_file, compression='gzip', nrows=None)

#Print value counts of the variable pregnum and compare results wtih NSFG codebook
print(respdf.pregnum.value_counts())

#Print pregnum as in NSFG codebook so that it's easy to compare
print(f'\n-----------------------\nNone {respdf[respdf.pregnum == 0].pregnum.count()}')
print(f'1 PREGNANCY {respdf[respdf.pregnum == 1].pregnum.count()}')
print(f'2 PREGNANCIES {respdf[respdf.pregnum == 2].pregnum.count()}')
print(f'3 PREGNANCIES {respdf[respdf.pregnum == 3].pregnum.count()}')
print(f'4 PREGNANCIES {respdf[respdf.pregnum == 4].pregnum.count()}')
print(f'5 PREGNANCIES {respdf[respdf.pregnum == 5].pregnum.count()}')
print(f'6 PREGNANCIES {respdf[respdf.pregnum == 6].pregnum.count()}')
print(f'7 OR MORE PREGNANCIES {respdf[respdf.pregnum >= 7].pregnum.count()}')


#Read pregnancy file using nsfg package method
pregdf = nsfg.ReadFemPreg()

# Create the map from caseid to list of pregnancy indices using method in nsfg package
preg_map = nsfg.MakePregMap(pregdf)

#Initialize a flag which we will use to confirm if recordset info matches
recordmatchFlag = True
#Now we iterate through respondent file pregnum series and if pregnum in respondent file does not match with 
#number of records in the pregnancy file, we break the loop and let the user know that something went wrong.
for index, pregnum in respdf.pregnum.items():
    caseid = respdf.caseid[index]
    indices = preg_map[caseid]
    #print(caseid, len(indices), pregnum)

    # check that pregnum from the respondent file equals
    # the number of records in the pregnancy file
    if len(indices) != pregnum:
        print(caseid, len(indices), pregnum)
        recordmatchFlag = False
        break
if recordmatchFlag:
    print("\nRecords matched between pregnancy and respondent datasets. Success")
else:
    print("\nRecords Mismatch between pregnancy and respondent datasets")
```

## Question 2 - 4
#### Using the variable totalwgt_lb, investigate whether first babies are lighter or heavier than others. Compute Cohen's d to quantify the difference between the groups. How does it compare to the difference in pregnancy length?

We will use effect size summary statistic to answer this question. Mean weight of first babies is 7.20 lb and of other babies is 7.32 lbs, a difference of .12 lbs. This difference is insignificant for the purposes of clearly identifying if first babies are lighter or heavier. 
We will now calculate Cohen's d which compares the difference between groups to the variability within groups. For Cohen's d general rule of thumb is
small effect = 0.2
medium effect = 0.5
large effect = 0.8
In our calculation we find that cohen's d is .08 therefore it's small effect.


```python
print(f'Mean weight of first born babies {firsts.totalwgt_lb.mean()}')
print(f'Mean weight of other babies {others.totalwgt_lb.mean()}')

diff_wt  = firsts.totalwgt_lb.mean() - others.totalwgt_lb.mean()
print(f'Difference in mean between first and other born babies {diff_wt}')
var1 = firsts.totalwgt_lb.var()
var2 = others.totalwgt_lb.var()
n1 = len(firsts.totalwgt_lb)
n2 = len(others.totalwgt_lb)
pooled_var = (n1 * var1 + n2 * var2) / (n1 + n2)
d = diff_wt / np.sqrt(pooled_var)
print(f'Cohens d {d}')
```
